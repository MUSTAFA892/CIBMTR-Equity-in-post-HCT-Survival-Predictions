```python


# Fill null values
for column in train_data.columns:
    if train_data[column].dtype == 'object':  # Categorical column
        train_data[column].fillna(train_data[column].mode()[0], inplace=True)
    else:  # Numerical column 
        train_data[column].fillna(train_data[column].mean(), inplace=True)
```

```python
train_data.drop(columns=['ID','efs_time'],inplace=True)
```

```python
categories = []
numerical = []

# Loop through all columns in the DataFrame
for col in train_data.columns:
    if train_data[col].dtype == 'object' or train_data[col].dtype.name == 'category':
        # If the column is categorical
        categories.append(col)
    else:
        # If the column is numerical
        numerical.append(col)

# Print the results
print("Categorical columns:", categories)
print("Number of categorical",len(categories))
print("Numerical columns:", numerical)
print("Number of numerical",len(numerical))
```

Developing a pipeline

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.preprocessing import FunctionTransformer

# Assuming train_data is already loaded as a DataFrame

# Identify categorical and numerical columns
categorical_columns = train_data.select_dtypes(include=['object', 'category']).columns
numerical_columns = train_data.select_dtypes(exclude=['object', 'category']).columns

# Exclude 'efs' from the numerical columns
numerical_columns = numerical_columns.drop('efs')

# Define the transformations for categorical columns
cat_pipeline = Pipeline([
    ('label_encoder', FunctionTransformer(lambda x: x.apply(LabelEncoder().fit_transform), validate=False))
])

# Commenting out the numerical scaling pipeline for now

num_pipeline = Pipeline([
    ('scaler', StandardScaler())
])

# Create a ColumnTransformer to apply different transformers to each column type
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', cat_pipeline, categorical_columns),
        # Commenting out the numerical transformer
        ('num', num_pipeline, numerical_columns)
    ])

# Apply the transformations
train_data_transformed = preprocessor.fit_transform(train_data)

# Since the ColumnTransformer outputs a numpy array, we need to get the original column names
# Create new column names based on the transformed columns
transformed_columns = list(categorical_columns)   + list(numerical_columns)# Fixing the list concatenation syntax

# Convert the numpy array back to a DataFrame
final_data = pd.DataFrame(train_data_transformed, columns=transformed_columns)
```
